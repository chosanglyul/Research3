```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 300, num=3000, endpoint=False)
y = np.sin(X)
```

```python
plt.plot(y)
```

```python
df = pd.DataFrame(y, index=X, columns=['y'])
for i in range(1, 200):
    df['X_{}'.format(i)] = df['y'].shift(i)
df.dropna(inplace=True)
df = df.loc[:, ::-1]
df
```

```python
df.to_csv('../preprocess/sincos.csv')
```
