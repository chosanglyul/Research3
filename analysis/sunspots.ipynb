```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../rawdata/sunspots.csv', index_col='Unnamed: 0')
df.columns = ['Date', 'Sunspot']
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df
```

```python
for i in range(1, 265):
    df['Sunspot_{}'.format(i)] = df['Sunspot'].shift(i)
df.dropna(inplace=True)
df = df.loc[:, ::-1]
df
```

```python
plt.plot(df['Sunspot'])
```

```python
a = ['Sunspot_{}'.format(i) for i in range(1, 265)]
a.append('Sunspot')
plt.plot(range(1, 266), df[a].corr()['Sunspot'])
```
